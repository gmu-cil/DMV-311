```python
#df_MC = pd.read_csv('C:\\Users\\samri\\Downloads\\311_Montgomery County_MD_2020.csv',index_col = 0)
df_MC = pd.read_csv('C:\\Users\\samri\\Downloads\\311_Montgomery Test.csv',index_col = 0)

from datetime import datetime, timedelta
df_MC["Date Service Due"] = pd.to_datetime(df_MC["Opened"])  + pd.to_timedelta(df_MC["Attached Solution SLA Days"],unit='d')
df_MC['Service Type'] = df_MC['Area'].fillna('') + ", " + df_MC['Sub-Area'].fillna('')
df_MC["Opened"] = pd.to_datetime(df_MC["Opened"])

df_MC["Closed"] = pd.to_datetime(df_MC["Closed"])
df_MC= df_MC.drop(columns = ['Congressional District', 'Congressional Member', 'Council District', 'Council Member Name','Changed Date','# of Days Open','SLA Yes', 'SLA No', 'Attached Solution SLA Days','Within SLA Windows' ,'opened_date','Source','Election District','Maryland State District','Area','Sub-Area'])
df_MC.rename(columns = {'SR ID': 'Service Request ID','Opened': 'Date Request Opened', 'Closed':'Date Request Resolved','Attached Solution (Topic)':'Service Description', 'Department':'Organization Responsible','Zip Code' :'ZipCode'},inplace = True)
df_MC = df_MC[['Service Request ID', 'Service Type', 'Service Description', 'Organization Responsible', 'Date Request Opened', 'Date Service Due', 'Date Request Resolved', 'Status','City', 'State', 'ZipCode']]

df_MC             
```

```python
import pandas as pd
import numpy as np
df_PG = pd.read_csv('C:\\Users\\samri\\Downloads\\311_Prince George County_MD_2019.csv',index_col = 0)
#df_PG = pd.read_csv('C:\\Users\\samri\\Downloads\\try_311_Prince George County_MD_2019.csv',index_col = 0)
df_PG= df_PG.drop(columns = ['SLA','Location','TNI Area Description', 'opened_date'])
df_PG['Latitude']= df_PG['Latitude'].astype(float)
df_PG['Longitude']= df_PG['Longitude'].astype(float)
df_PG['Latitude'] = df_PG['Latitude'].replace(np.nan, 0, regex=True)
df_PG['Longitude'] = df_PG['Longitude'].replace(np.nan, 0, regex=True)
df_PG["Date Request Opened"] = pd.to_datetime(df_PG["Date Request Opened"])
df_PG.rename(columns = {'Request Name' : 'Service Description', 'Service Group' : 'Service Type', 'Agency Responsible' : 'Organization Responsible', 'Request Status' : 'Status', 'Service Request #' : 'Service Request ID'}, inplace = True)
df_PG = df_PG[['Service Request ID', 'Service Type', 'Service Description', 'Organization Responsible', 'Date Request Opened','Status','Latitude', 'Longitude']]
df_PG.to_csv(r'C:\\Users\\samri\\OneDrive\\Documents\\Research work\\DATA\\Formatted_311_Prince George County_MD_2019.csv', index = False)
#df_PG.to_csv(r'C:\\Users\\samri\\OneDrive\\Documents\\Research work\\DATA\\try_311_Prince George County_MD_2019.csv', index = False)

df_PG
```
