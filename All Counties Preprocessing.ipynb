#### Montgomery County Data Preprocessing

```python
import pandas as pd
df_MC = pd.read_csv('C:\\Users\\samri\\OneDrive\\Documents\\Research work\\DATA\\Montgomery County\\Montgomery County\\311_Montgomery County_MD_2012_2021.csv')
```

1. Calculate 'Date Service Due' using 'Opened' and 'Attached Solution SLA Days'
2. Convert 'Opened' and 'Date Service Due' to data time
3. Merge 'Area' and 'Sub-area' column to get 'Service Type'
4. Rename the columns as per unified database


```python
from datetime import datetime, timedelta
df_MC["Date Service Due"] = pd.to_datetime(df_MC["Opened"])  + pd.to_timedelta(df_MC["Attached Solution SLA Days"],unit='d')
df_MC['Service Type'] = df_MC['Area'].fillna('') + ", " + df_MC['Sub-Area'].fillna('')
df_MC["Opened"] = pd.to_datetime(df_MC["Opened"])

df_MC.rename(columns = {'SR ID': 'Service Request ID','Opened': 'Date Request Opened', 'Closed':'Date Request Resolved','Attached Solution (Topic)':'Service Description', 'Department':'Organization Responsible','Zip Code' :'ZipCode'},inplace = True)
df_MC["Details"]=""
df_MC["Longitude"]=""
df_MC["Latitude"]=""
df_MC["Street Address"]=""
df_MC = df_MC[['Service Request ID', 'Service Type', 'Service Description', 'Organization Responsible', 'Date Request Opened', 'Date Service Due', 'Date Request Resolved', 'Status','Details','Street Address','City', 'State', 'ZipCode', 'Latitude','Longitude']]            
```

```python
df_MC.to_csv(r'C:\\Users\\samri\\OneDrive\\Documents\\Research work\\DATA\\_Formatted_311_Montgomery County_MD_2012_2021.csv', index = False)
```

#### PG County Data Preprocessing

```python
import pandas as pd
import numpy as np
df_PG = pd.read_csv('C:\\Users\\samri\\OneDrive\\Documents\\Research work\\DATA\\Prince George County\\Prince George County\\311_Prince George County_MD_2019.csv',index_col = 0)
```

1. Drop unnecessary columns
2. Change lat/ lon fields to float and update nan values to zeros
3. Change 'Date Request Opened' to date time
4. Rename the columns according to unified database

```python
df_PG= df_PG.drop(columns = ['SLA','Location','TNI Area Description', 'opened_date'])
df_PG['Latitude']= df_PG['Latitude'].astype(float)
df_PG['Longitude']= df_PG['Longitude'].astype(float)
df_PG['Latitude'] = df_PG['Latitude'].replace(np.nan, 0, regex=True)
df_PG['Longitude'] = df_PG['Longitude'].replace(np.nan, 0, regex=True)
df_PG["Date Request Opened"] = pd.to_datetime(df_PG["Date Request Opened"])
df_PG["Date Service Due"]=""
df_PG["Date Request Resolved"]=""
df_PG["Details"]=""
df_PG["ZipCode"]=""
df_PG["City"]=""
df_PG["State"]=""
df_PG.rename(columns = {'Request Name' : 'Service Description', 'Service Group' : 'Service Type', 'Agency Responsible' : 'Organization Responsible', 'Request Status' : 'Status', 'Service Request #' : 'Service Request ID'}, inplace = True)
df_PG = df_PG[['Service Request ID', 'Service Type', 'Service Description', 'Organization Responsible', 'Date Request Opened', 'Date Service Due', 'Date Request Resolved', 'Status','Details','Street Address','City', 'State', 'ZipCode', 'Latitude','Longitude']]
```

#### Arlington County Data Preprocessing

```python
df_AC = pd.read_csv('C:\\Users\samri\\OneDrive\\Documents\\Research work\\DATA\\Arlington County\\Arlington County\\MyArlington_Arlington_VA_2020_2021.csv',index_col = 0,na_filter = False)
```

1. Drop unnecessary columns
2. Convert 'opened_date', 'updated_date' to date time
3. Rename columns as per unified database


```python
df_AC= df_AC.drop(columns = ['service_request_key', 'permission'])
df_AC["opened_date"] = pd.to_datetime(df_AC["created_at"])
df_AC["updated_date"] = pd.to_datetime(df_AC["updated_at"])
df_AC['latitude'] = pd.to_numeric(df_AC['latitude'])
df_AC['longitude'] = pd.to_numeric(df_AC['longitude'])
df_AC['Date Service Due'] =""
df_AC["City"]=""
df_AC["Date Request Resolved"]=""
df_AC["Organization Responsible"]=""
df_AC["State"]=""
df_AC["ZipCode"]=""
df_AC['latitude']=df_AC['latitude'].values.astype(float)
df_AC['longitude']= df_AC['longitude'].values.astype(float)
df_AC['latitude'] = df_AC['latitude'].replace(np.nan, 0, regex=True)
df_AC['longitude'] = df_AC['longitude'].replace(np.nan, 0, regex=True)
df_AC.rename(columns = {'service_request_id': 'Service Request ID','name':'Service Type','description' : 'Service Description',  'created_at' : 'Date Request Opened', 'updated_at': 'Date Service Due','status' : 'Status', 'address' : 'Street Address', 'latitude' : 'Latitude', 'longitude': 'Longitude', 'source_system':'Details' }, inplace = True)
df_AC = df_AC[['Service Request ID', 'Service Type', 'Service Description', 'Organization Responsible', 'Date Request Opened', 'Date Service Due', 'Date Request Resolved', 'Status', 'Details','Street Address', 'City', 'State', 'ZipCode', 'Latitude', 'Longitude']]
```
